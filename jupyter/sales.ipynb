```python
import pandas as pd
import numpy as np

df_customer_data = pd.read_csv('../data-source/2021/2nd term/customer-export-10012022-072818-61dbe012.csv')
df_customer_item_details = pd.read_csv('../data-source/2021/2nd term/Report Item Details - 26-04-2021 - 25-12-2021 - Outlet_Sitisonya - 088f3fbb.csv')
df_customer_transactions = pd.read_csv('../data-source/2021/2nd term/Report Transaction - 26-04-2021 - 25-12-2021 - Outlet_Sitisonya - c0929211.csv')

```

Adjust data type

```python
# Customer Data
df_customer_data['Member Since'] = pd.to_datetime(df_customer_data['Member Since'],format='%d-%m-%Y')
df_customer_data['Last Visit'] = pd.to_datetime(df_customer_data['Last Visit'],format='%d-%m-%Y')
df_customer_data['revisit'] = np.where(df_customer_data['Member Since'] == df_customer_data['Last Visit'],True,False )

#Items Details
df_customer_item_details['Customer'] = pd.to_numeric(df_customer_item_details['Customer'],errors='coerce')

#Transactions
df_customer_transactions['Date'] = pd.to_datetime(df_customer_transactions['Date'])
df_customer_transactions['Day'] = df_customer_transactions['Date'].dt.day_name()
```

```python
grouped_period_sex =  df_customer_data.groupby([df_customer_data['Member Since'].dt.year.rename('year'), df_customer_data['Member Since'].dt.month.rename('month'), df_customer_data['Sex'],df_customer_data['revisit']],dropna=False).agg({'count'})
grouped_period_sex['Phone']
```

```python
grouped_transaction =  df_customer_transactions.groupby([df_customer_transactions['Customer Phone']]).agg({'count'})
grouped_transaction['Receipt Number'].sum()
```

Item Details

```python
#pgm = (df_customer_data['Member Since'] >= pd.to_datetime('12-4-2021')) & (df_customer_data['Member Since'] <= pd.to_datetime('12-5-2021'))
#pgm_event_revisit =  df_customer_data.loc[pgm].groupby([df_customer_data['Sex'],df_customer_data['revisit']],dropna=False).agg({'count'})
#pgm_event_revisit
weekend = (df_customer_transactions['Date'] >= pd.to_datetime('1-11-2021')) & (df_customer_transactions['Date'] <= pd.to_datetime('31-12-2021')) & (df_customer_transactions['Date'].dt.dayofweek > 4)
sorting_weekend =  df_customer_transactions[weekend].groupby(['Date']).sum()
sorting_weekend.to_excel('../result/2021/2nd term/weekend.xlsx')
sorting_weekend
```

```python
first_ads = (df_customer_transactions['Date'] >= pd.to_datetime('6-1-2021')) & (df_customer_transactions['Date'] <= pd.to_datetime('6-22-2021')) 
sorting_first_ads =  df_customer_transactions[first_ads].groupby(['Date']).sum()
sorting_first_ads.to_csv('../result/2021/2nd term/sorting_first_ads.csv')
sorting_first_ads
```
